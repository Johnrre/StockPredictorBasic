```python
import yfinance as yf
```

```python
stocks = yf.Ticker("^NDX")
stocks
```

```python
stocks_history = stocks.history(period = 'max')
stocks_history
```

```python
stocks_history = stocks_history.loc['1990-01-01':].copy()
stocks_history
```

```python
del stocks_history['Dividends']
del stocks_history['Stock Splits']
```

```python
stocks_history.plot.line(y = "Close", use_index = True)
```

```python
stocks_history["Tomorrow's_Price"] = stocks_history['Close'].shift(-1)
stocks_history
```

```python
stocks_history['Conclusion'] = (stocks_history["Tomorrow's_Price"] > stocks_history["Close"]).astype(int)
stocks_history
```

```python
from sklearn.ensemble import RandomForestClassifier #we use this one as its harder to overfit this type of model also it picks up non linearity in the data

model = RandomForestClassifier(n_estimators=100, min_samples_split = 100, random_state=1)  # n is the number of decision trees, min_samples == helps in overfitting (higher means less overfiting but less accuracy) random_state just makes sure the randomness is predictable so that we get the 'same-ish output' every time we run

training_set = stocks_history.iloc[:-100]
test_set = stocks_history.iloc[-100:]

prediction_parameters = ['Close', "Volume", 'Open', 'High', 'Low']
model.fit(training_set[prediction_parameters], training_set['Conclusion'])
```

```python
from sklearn.metrics import precision_score  #basically gives the percent of correct outputs

preds = model.predict(test_set[prediction_parameters])
```

```python
preds
```

```python
import pandas as pd
predictions = pd.Series(preds, index = test_set.index)
```

```python
predictions
```

```python
precision_score(test_set['Conclusion'], predictions)  #intial output of approx 60% accuracy not bad but can do better
```

```python
combined = pd.concat([test_set['Conclusion'], predictions], axis = 1)
```

```python
combined.plot()
```

```python
def predicting(train, test, predictors, model):
  model.fit(train[predictors], train['Conclusion'])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds, index = test.index, name = 'Predictions')
  combined = pd.concat([test['Conclusion'], preds], axis = 1)
  return combined
```

```python
def backtest(data, model, predictors, start = 2500, step = 250): # 1 yr has 250 stock days so 2500 is 10 yrs of data, step is 1 yr of data
  all_predictions = []

  for i in range(start, data.shape[0], step):
    training_set = data.iloc[0:i].copy()
    test_set = data.iloc[i:(i+step)].copy()
    predictions = predicting(training_set, test_set, predictors, model)
    all_predictions.append(predictions)

  return pd.concat(all_predictions)
```

```python
new_predictions = backtest(stocks_history, model, prediction_parameters)
```

```python
new_predictions['Predictions'].value_counts()
```

```python
precision_score(new_predictions['Conclusion'], new_predictions['Predictions'])
```

```python
new_predictions['Conclusion'].value_counts() / new_predictions.shape[0]
```

```python
days = [2,5,60,250,1000]    #Logically a normal person would use data from 2 days ago, 5 days ago, 3 months ago..... to predict tommorrows stock movement
new_predictors = []

for horizon in days:
    rolling_averages = stocks_history.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    stocks_history[ratio_column] = stocks_history["Close"] / rolling_averages["Close"]   #Adding more 'predictors'  (current row)/(average of previous row)

    trend_column = f"Trend_{horizon}"
    stocks_history[trend_column] = stocks_history.shift(1).rolling(horizon).sum()["Conclusion"]  #Adding rolling sum of current conclusion/target

    new_predictors+= [ratio_column, trend_column]

stocks_history = stocks_history.dropna(subset=stocks_history.columns[stocks_history.columns != "Tomorrow"])
stocks_history
```

```python
new_model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.65] = 1
    preds[preds <.65] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(stocks_history, new_model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Conclusion"], predictions["Predictions"])
```
